```python
import matplotlib.pyplot as plt
%matplotlib inline
```

# Pie Chart

```python
areas = ['Marketing','Sales','Development','HR','Customer Support']
budget = [2,3,6,5,7]
```

```python
plt.pie(budget, labels=areas)
```

```python
plt.pie(budget, labels=areas)
plt.show()
```

```python
plt.pie(budget, labels=areas, autopct="%0.1f%%")
plt.show()
```

```python
plt.pie(budget, labels=areas, autopct="%0.2f%%")
plt.show()
```

```python
x = [11,44,66,33,88,99]
y = [2,5,8,4,9,1]
plt.scatter(x,y)
plt.xlabel("Age")
plt.ylabel("Salary")
plt.title("Age vs Salary")
```

```python
pid = [1,2,3,4,5,6,7,8]
onp = [333,788,100,677,455,800,300,233]
```

```python
plt.plot(pid,onp,color="red", marker="+", markersize=10)
```

```python
ofp = [220,550,400,700,800,122,500,344]
cb=[100,200,300,400,150,600,400,800]
```

```python
plt.plot(pid,onp,color="red", marker="+", markersize=10, label="Online")
plt.plot(pid,ofp,color="blue", marker="D", markersize=10, label = "Offline")
plt.plot(pid,cb,color="green", markersize=10, label="Chor Bazaar")
plt.legend()
plt.show()
```

# Bar graph

```python
year=[2010,2011,2012,2013,2014,2015]
price = [100,200,300,350,300,500]
cprice=[900,500,300,600,400,200]
```

```python
plt.bar(year,price,color="red")
plt.bar(year,cprice,color="blue")
```

```python
plt.barh(year,price,color="red")
```
