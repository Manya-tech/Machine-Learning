Loading Dataset

```python
import pandas as pd
```

```python
csgo=pd.read_csv("csgo_round_snapshots.csv")
csgo.head()
```

```python
csgo.info()
```

```python
csgo.isna().sum()
```

```python
df=pd.get_dummies(csgo,drop_first=True)
```

```python
df
```

```python
from sklearn.preprocessing import StandardScaler   #scaling data

scaler=StandardScaler()

array=scaler.fit_transform(df.drop("round_winner_T",axis=1))

dif=pd.DataFrame(array,columns=df.columns[:-1])
```

```python
dif.head()
```

```python
x=dif
y=df["round_winner_T"]
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=100)  #splitting dataset
```

KNN Model

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import GridSearchCV

knc=KNeighborsClassifier()

k_range=list(range(1,6))
param_grid=dict(n_neighbors=k_range)

grid=GridSearchCV(knc,param_grid,cv=10,scoring='accuracy',return_train_score=False,verbose=1)

grid_search=grid.fit(x_train,y_train)
```

```python
grid_search.best_params_
```

```python
acc=grid_search.best_score_*100
print("Accuracy for our training dataset with tuning is",acc,"%")
```

```python
knn=KNeighborsClassifier(n_neighbors=1)

knn.fit(x_train,y_train)

predict=knn.predict(x_test)
```

Checking Accuracy

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_confusion_matrix

matrix=confusion_matrix(y_test,predict)
matrix
```

```python
plot_confusion_matrix(grid,x_test,y_test,values_format='d')
```

```python
print(classification_report(y_test,predict))
```
