```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
cancer.target
```

```python
cancer.target_names
```

```python
df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
dfs=scaler.fit_transform(df)
dfs
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(dfs,cancer['target'],test_size=0.33,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=2)
```

```python
pca.fit(dfs)
```

```python
d_pca = pca.transform(dfs)
```

```python
d_pca.shape
```

```python
x_train,x_test,y_train,y_test = train_test_split(d_pca,cancer['target'],test_size=0.33,random_state=42)
```

```python
lr = LogisticRegression()
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
plt.figure(figsize=(8,6))
plt.scatter(d_pca[:,0],d_pca[:,1],c=cancer['target'])
plt.xlabel("Principal Component 1")
plt.ylabel("principal Component 2")
```
