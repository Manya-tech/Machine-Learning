```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
dfs=scaler.fit_transform(df)
dfs
```

```python
y = cancer['target']
```

```python
y
```

```python
features = dfs.T
print(features)
cov_matrix = np.cov(features)
```

```python
cov_matrix
```

```python
cov_matrix.shape
```

```python
values,vectors = np.linalg.eig(cov_matrix)
values
```

```python
vectors
```

```python
percentage = []
for value in values:
    percentage.append((value/np.sum(values))*100)
```

```python
percentage
```

```python
d1 = np.dot(dfs,vectors.T[0])
d2 = np.dot(dfs,vectors.T[1],)
d3 = np.dot(dfs,vectors.T[2],)
```

```python
result = pd.DataFrame(d1,columns=['PC1'])
```

```python
result['PC2']=d2
result['PC3']=d3
```

```python
result
```

```python
result['target']=y
```

```python
result
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
plt.figure(figsize=(20,10))
sns.scatterplot(x=result['PC1'],y=result['PC2'],hue=result['target'])
```

```python
new = result.drop(columns=['target'],axis=1)
```

```python
new
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(new,y,test_size=0.33,random_state=42)
```

```python
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```
