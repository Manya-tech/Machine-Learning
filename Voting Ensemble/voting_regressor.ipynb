```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('BostonHousing.csv')
```

```python
df.head()
```

```python
x=df.iloc[:,0:13]
y=df.iloc[:,-1]
x
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score
```

```python
lr=LinearRegression()
dt=DecisionTreeRegressor()
svr=SVR()
```

```python
estimators=[('lr',lr),('dt',dt),('svr',svr)]
for estimator in estimators:
    s=cross_val_score(estimator[1],x,y,scoring='r2',cv=10)
    print(estimator[0],np.round(np.mean(s),2))
```

```python
from sklearn.ensemble import VotingRegressor
```

```python
vr=VotingRegressor(estimators=estimators)
s=cross_val_score(vr,x,y,scoring='r2',cv=10)
print('VR',np.round(np.mean(s),2))
```

```python
for i in range(1,4):
    for j in range(1,4):
        for k in range(1,4):
            vr = VotingRegressor(estimators=estimators,weights=[i,j,k])
            s=cross_val_score(vr,x,y,scoring='r2',cv=10)
            print("for i={},j={},k={}".format(i,j,k),np.round(np.mean(s),2))
```

```python
# using same algorithm

d1=DecisionTreeRegressor(max_depth=1)
d2=DecisionTreeRegressor(max_depth=3)
d3=DecisionTreeRegressor(max_depth=5)
d4=DecisionTreeRegressor(max_depth=7)
d5=DecisionTreeRegressor(max_depth=None)
```

```python
estimators=[('d1',d1),('d2',d2),('d3',d3),('d4',d4),('d5',d5)]
```

```python
for estimator in estimators:
    s=cross_val_score(estimator[1],x,y,scoring='r2',cv=10)
    print(estimator[0],np.round(np.mean(s),2))
```

```python
vr=VotingRegressor(estimators=estimators,verbose=True)
s=cross_val_score(vr,x,y,scoring='r2',cv=10)
print('VR',np.round(np.mean(s),2))
```
