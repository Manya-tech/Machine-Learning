```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df= pd.read_csv("Iris.csv")
df = df.rename(columns={"SepalLengthCm":"sl","SepalWidthCm":"sw","PetalLengthCm":"pl","PetalWidthCm":"pw"})
s = df[df['Species']=='Iris-setosa']
ver = df[df['Species']=='Iris-versicolor']
vir = df[df['Species']=='Iris-virginica']
```

```python
sns.distplot(s['pl'])
```

```python
sns.FacetGrid(df, hue='Species',height=6).map(sns.distplot,'pl').add_legend()
```

```python
sns.FacetGrid(df, hue='Species',height=6).map(sns.distplot,'pl',kde=False).add_legend()
```

```python
sns.FacetGrid(df, hue='Species',height=6).map(sns.distplot,'pw').add_legend()
```

# calculate pdf and cdf

```python
counts, bin_edges=np.histogram(s['pl'],bins=10)
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
counts, bin_edges=np.histogram(ver['pl'],bins=10)
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
counts, bin_edges=np.histogram(vir['pl'],bins=10)
pdf = counts/sum(counts)
cdf = np.cumsum(pdf)
plt.plot(bin_edges[1:],pdf)
plt.plot(bin_edges[1:],cdf)
```
