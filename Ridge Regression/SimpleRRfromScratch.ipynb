```python
from sklearn.datasets import make_regression
import matplotlib.pyplot as plt
import numpy as np
X,y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1,noise=20,random_state=13)
plt.scatter(X,y)
```

```python
from sklearn.linear_model import LinearRegression, Ridge

lr = LinearRegression()
lr.fit(X,y)
print(lr.coef_)
print(lr.intercept_)
```

```python
rr = Ridge(alpha=10)
rr.fit(X,y)
print(rr.coef_)
print(rr.intercept_)
```

```python
rr1 = Ridge(alpha=100)
rr1.fit(X,y)
print(rr1.coef_)
print(rr1.intercept_)
```

```python
plt.plot(X,lr.predict(X), color='red', label='aplha=0')
plt.plot(X,rr.predict(X), color='blue', label='aplha=10')
plt.plot(X,rr1.predict(X), color='green', label='aplha=100')
plt.scatter(X,y)
plt.legend()
```
