# filtering df on condition (imp)

use & and | because bitwise operators used with booleans

then add in df

```python
mask1 = df['total_rooms']>1000
mask1
```

```python
df[mask1].head(10)
```

```python
mask2 = df['total_bedrooms']>300
```
