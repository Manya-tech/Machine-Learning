# is_in()

```python
df.head()
```

```python
df.info()
```

```python
do = df[df['over']>15]
```

```python
bm = do.groupby("batsman")
```

```python
all_bm=bm['batsman_runs'].count()
```

```python
x= all_bm>200
```

```python
x
```

```python
bm_list=all_bm[x].index.to_list()
```

```python
len(bm_list)
```

```python
final = df[df['batsman'].isin(bm_list)]
```

```python
final
```

```python
runs = final.groupby("batsman")['batsman_runs'].sum()
```

```python
balls = final.groupby("batsman")['batsman_runs'].count()
```

```python
strike_rate = (runs*100)/balls
```

```python
strike_rate
```
